```python
import csv

image_file = "/kaggle/working/HARRISON/data_list.txt"  # Path to the image entries text file
text_file = "/kaggle/working/HARRISON/tag_list.txt"  # Path to the text entries text file
output_file = "/kaggle/working/output.csv"  # Path to the output CSV file

image_data = {}
with open(image_file, 'r') as file:
    img_num = 1
    for line in file:
        path = line.strip()
        image_data[img_num] = path
        img_num += 1

text_data = []
tag_dict = dict()
with open(text_file, 'r') as file:
    img_num = 1
    tag_num = 0
    for line in file:
        text = line.strip()
        tags = line.split(" ")
        for tag in tags:
          if tag not in tag_dict.keys():
            tag_dict[tag] = tag_num
            tag_num += 1
        path = image_data.get(img_num, '')
        text_data.append([img_num, path, text])
        img_num += 1

with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Img_num', 'Path', 'Text'])
    writer.writerows(text_data)

print("CSV file created successfully!")
```

```python
import torch
import pandas as pd
from typing import Dict
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import matplotlib.pyplot as plt
from PIL import Image

class HashtagDataset(Dataset):
  def __init__(self, data_dir: str, data_csv: str, tag_dict: Dict, img_size) -> None:
    self.data_dir = data_dir
    self.data_csv = data_csv
    self.df = pd.read_csv(data_csv)
    self.df = self.df.sample(frac=0.1)
    self.tag_dict = tag_dict
    self.img_size = img_size

  def __len__(self):
    return len(self.df.index)

  def __getitem__(self, idx):
    record = self.df.iloc[idx, :]
    img_num, path, text = record
    
    img = Image.open(str(self.data_dir + path))
    transforms = T.Compose([
        T.Resize(size=self.img_size),
        T.ToTensor()
    ])
    img  = transforms(img)
    if img.shape[0] == 1:
      img = torch.cat([img, img, img], dim=0)
    
    tags_list = text.split(" ")
    tags_list = [tag_dict[tag] for tag in tags_list]
    label = torch.zeros(size=(998,))
    for tag in tags_list:
      label[tag] = 1
    return {"img": img, "label": label}
```

```python
def collate_fn(batch):
  return {
      "img": torch.stack([x["img"] for x in batch]),
      "label": torch.stack([x["label"] for x in batch])
  }
```

```python
from transformers import ViTForImageClassification
model_name = "google/vit-base-patch16-224-in21k"
model = ViTForImageClassification.from_pretrained(
    model_name,
    num_labels=998   
)

```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)
```

```python
from transformers import get_scheduler
from torch.optim import AdamW
optimizer = AdamW(model.parameters(), lr=1e-4)
num_epochs = 5
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)
```

```python
t = model(hashtag_dataset[0]["img"].unsqueeze(dim=0).to(device))
t.__dir__()
```

```python
from torch import nn

class ViTClassifier(nn.Module):
  def __init__(self, model):
    super(ViTClassifier, self).__init__()
    self.vit = model
    
  def forward(self, img):
    vit_image_classification_output = self.vit(img)
    logits = vit_image_classification_output.logits
    return logits

model = ViTClassifier(model)
```

```python
from tqdm.auto import tqdm
progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
  for i, batch in enumerate(train_dataloader):
    batch = {k: v.to(device) for k, v in batch.items()}
    outputs = model(batch["img"])
    # print(outputs.shape, batch["label"].shape)
    criterion = torch.nn.CrossEntropyLoss()
    loss = criterion(outputs, batch["label"].float())
    loss.backward()
    optimizer.step()
    lr_scheduler.step()
    optimizer.zero_grad()
    progress_bar.update(1)
    writer.add_scalar('training_loss', loss.item(), epoch*len(train_dataloader) + i+1)

  print(f"EPOCH {epoch+1}/{num_epochs}, after current batch, Loss: {loss}")
```

```python
import torch
import torchvision.transforms as transforms
import gradio as gr

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model_new.to(device)
model_new.eval()

def predict_hashtags(img, topk=10):
  transforms_list = transforms.Compose([
      transforms.Resize(size=(224, 224)),
      transforms.ToTensor()
  ])
  img = transforms_list(img)
  if img.shape[0] == 4:
    img = img[:3, :, :]
  img = img.unsqueeze(dim=0).to(device)
  with torch.no_grad():
    logits = model_new(img)
  scores, tags_present = logits.squeeze().topk(10)
  tags = ""
  for ind in tags_present:
    tags += "#" + str(ind_to_tag[ind.item()]) + " "
  return tags

gr.Interface(
    fn=predict_hashtags,
    inputs=gr.Image(type="pil"),
    outputs=gr.Textbox(lines=10)
).launch(share=True)
```
